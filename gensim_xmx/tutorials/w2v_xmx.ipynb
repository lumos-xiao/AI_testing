```python
from gensim import corpora


# collect statistics about all tokens
dictionary = corpora.Dictionary(line.lower().split() for line in open('/home/xmx/projects/AI_testing/gensim_xmx/mycorpus.txt'))
# remove stop words and words that appear only once
stoplist = set('for a of the and to in'.split())
stop_ids = [
    dictionary.token2id[stopword]
    for stopword in stoplist
    if stopword in dictionary.token2id
]
once_ids = [tokenid for tokenid, docfreq in dictionary.dfs.items() if docfreq == 1]
dictionary.filter_tokens(stop_ids + once_ids)  # remove stop words and words that appear only once
dictionary.compactify()  # remove gaps in id sequence after words that were removed
print(dictionary)
```

```python
from gensim.test.utils import datapath
from gensim import utils

class MyCorpus:
    """An iterator that yields sentences (lists of str)."""

    def __iter__(self):
        corpus_path = datapath('/home/xmx/projects/AI_testing/gensim_xmx/mycorpus.txt')
        for line in open(corpus_path):
            # assume there's one document per line, tokens separated by whitespace
            yield utils.simple_preprocess(line)
```

```python
sentences = MyCorpus()
for vector in sentences:  # load one vector into memory at a time
    print(vector)
```

```python
import gensim.models

sentences = MyCorpus()
model = gensim.models.Word2Vec(sentences=sentences)
```

```python
for index, word in enumerate(model.wv.index_to_key):
    if index == 10:
        break
    print(f"word #{index}/{len(model.wv.index_to_key)} is {word}")
```
